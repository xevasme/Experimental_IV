## 1.1 Modelo Normal

### Experimento 1: Medici√≥n de alturas en una poblaci√≥n
‚Ä¢ **Material**: Cinta m√©trica o regla, cuaderno para registro.

‚Ä¢ **Procedimiento**:
1. Seleccionar un grupo de personas de una poblaci√≥n determinada.
2. Medir sus alturas y registrar los valores.
3. Calcular la media y la desviaci√≥n est√°ndar de las alturas.
4. Graficar la distribuci√≥n de las alturas para observar si se ajusta a una distribuci√≥n normal.


```python
# Simulaci√≥n de alturas (en cm) - reemplazar con tus datos reales si los tienes
alturas = np.random.normal(loc=170, scale=7, size=50)  # media = 170 cm, œÉ = 7 cm

# C√°lculo de estad√≠sticos
media = np.mean(alturas)
desviacion = np.std(alturas, ddof=1)

print(f"Media: {media:.2f} cm")
print(f"Desviaci√≥n est√°ndar: {desviacion:.2f} cm")

# Histograma
plt.figure(figsize=(8, 5))
count, bins, ignored = plt.hist(alturas, bins=10, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Curva de distribuci√≥n normal ajustada
x = np.linspace(min(alturas), max(alturas), len(alturas))
pdf = norm.pdf(x, media, desviacion)
plt.plot(x, pdf, 'r--', label='Distribuci√≥n Normal Ajustada')

plt.title("Distribuci√≥n de Alturas")
plt.xlabel("Altura (cm)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
